###                        Aqui se mostraran el proceso que se utilizo para limpiar los datos y comprenderlos mejor

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos
ruta = "\\Users\\TB\\Downloads\\StockMarketDataset.csv"
df = pd.read_csv(ruta)
df = df.rename(columns={'Unnamed: 0': 'ID'})
df.head(3)
```

```python
df_head = df.head()
df_info = df.info()
df_describe = df.describe()

print(df_head, df_info, df_describe)
```

```python
#Establecer la fecha
df.set_index('Date', inplace=True)
# Convertir valores a flotantes
for col in df.columns:
    df[col] = df[col].replace(',', '', regex=True).astype('float64')
```

```python
#Que columnas tienen datos nulos?
df.isnull().sum()
```

```python
#Remplazo los datos nulos con el valor a su lado 
df.bfill(inplace=True)
df.isnull().sum()
```

```python
#Limpieza de datos finalzada con EXITO!
df.to_csv('datos_limpios_mercado.csv', index=True)
```
